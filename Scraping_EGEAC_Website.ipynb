```python

# Import the two main libraries
import urllib.request               # to open the web pages
from bs4 import BeautifulSoup       # to process html
import ssl                          # to process ssl certificates
import pandas as pd
from datetime import date
import regex as re
from googletrans import Translator, LANGUAGES
```

```python
# Allow not verified SSL (Secure Socket Layer) certificates to be open
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# Open a Tripadvisor web page (simulating that is a Safari browser - Tripadvisor does not allow bots no open pages - more on agents: https://developers.whatismybrowser.com)
url = "https://egeac.pt/programacao-espacos-culturais/"
req = urllib.request.Request(
    url, 
    data=None, 
    headers={
        'User-Agent': 'https://developers.whatismybrowser.com/useragents/parse/1274624safari-macos-webkit'
    }
)
page = urllib.request.urlopen(req, timeout=60)
```

```python
# Read html and decode it to the UTF-8 encoding as a string
html = page.read().decode("utf-8")  # Can generate errors if time-out
```

```python
# Parse the html into a beatuful soup object
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
all_places = []
all_categories = []
all_spectacles = []
all_dates = []

categories = soup.find_all("p", {"class": "category"})

for category in categories:
    if category:
        category_text = category.get_text().strip()
        all_categories.append(category_text)
    else:
        all_categories.append("category not found")


places = soup.find_all("div", {"class": "location"})


for place in places:
    if place:
        place_text = place.get_text().strip()
        all_places.append(place_text)
    else:
        all_places.append("place not found")

spectacles = soup.find_all("p", {"class": "h5"})

for spectacle in spectacles:
    if spectacle:
        spectacle_text = spectacle.get_text().strip()
        all_spectacles.append(spectacle_text)
    else:
        all_spectacles.append("spectable not found")


dates = soup.find_all("div", {"class": "date"})

for date in dates:
    if date:
        date_text = date.get_text().split()
        all_dates.append(date_text)
    else:
        all_dates.append("date not found")

```
